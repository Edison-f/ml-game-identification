```python
import os
import torch
from PIL import Image
```

## Model Input Sizes

```python
width = 640
height = 360

screenshot_frequency = 30 # How often, in seconds, to take a screenshot
```

## Process Images

```python
for file in os.listdir(os.path.join('input', 'images')):
    output_path = os.path.join('output', '.'.join(file.split('.')[:-1]) + '.png')
    Image.open(os.path.join('input', 'images', file)).resize(size = (width, height)).save(fp = output_path, format = 'png')
```

## Process Videos

```python
for file in os.listdir(os.path.join('input', 'video')): # Tested using ffmpeg version 2021-12-20-git-631e31773b-full_build-www.gyan.dev
    input_path = os.path.join('input', 'video', file)
    output_path = os.path.join('output', '.'.join(file.split('.')[:-1]) + '_%06d.png')
    rate = 1 / screenshot_frequency
    command = ['ffmpeg', f'-i \"{input_path}\"', f'-r {rate}', f'\"{output_path}\"', f'-s {width}x{height}', '-ss 00:30:00']
    if os.system(' '.join(command)) != 0:
        print(f'Error at {file}')
```

```python
for file_path in open(os.path.join('input', 'video_files.txt')): # Read video files from file paths
    file = os.path.split(file_path)[1]
    output_path = os.path.join('output', '.'.join(file.split('.')[:-1]) + '_%06d.png')
    rate = 1 / screenshot_frequency
    command = ['ffmpeg', f'-i \"{file_path}\"', f'-r {rate}', f'\"{output_path}\"', f'-s {width}x{height}', '-ss 00:30:00']
    if os.system(' '.join(command)) != 0:
        print(f'Error at {file}')
```
