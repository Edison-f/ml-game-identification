```python
import os
from PIL import Image
```

## Model Input Sizes

```python
width = 1280
height = 720
```

## Process Images

```python
# Iterate over each file in input
for file in os.listdir('input'):
    print(file)
    # Normalize it to the correct dimensions using PIL
    output_path = os.path.join('output', '.'.join(file.split('.')[:-1]) + '.png')
    print(output_path)
    Image.open(os.path.join('input', file)).resize(size = (1280, 720)).save(fp = output_path, format = 'png')
```
